#### 4. Observa cuántos Nas tienes por columna y por fila (usa la función que vimos para hacerlo para todas las columnas/filas de una vez). Elimina las filas en donde haya valores nulos.

```python
import numpy as np

# Función vista en clase
def num_nulos(x):
  return x.isnull().sum()

print("Número de Nas por fila: ")
print("Nas por fila:")
print(train_set.apply(num_nulos, axis=1))

print("\n")
print("Número de Nas por columna: ")
print(train_set.apply(num_nulos, axis=0))

# Para eliminar las filas con Nas, agregamos una columna que nos diga si la fila tiene nulos

train_set.loc[:, 'n_nulos'] = train_set.apply(num_nulos, axis=1)
train_set.info()
print( "Registros sin nulos: " + str( len(train_set[train_set['n_nulos'] == 0 ])))

train_set = train_set[train_set.n_nulos == 0 ]

# Validando eliminar filas con Nas
print("\n")
train_set.info()
```

#### 5. Recodifica a 0-1 las variables categóricas que estén siendo tratadas como strings. Usa la función `replace` que vimos, con un diccionario para hacer esto.

#### 7. Para las variables categóricas, usa una `crosstab` para saber si se les da más créditos a las mujeres que a los hombres.  Usa otra `crosstab` para saber si el hecho de tener historial crediticio hace que sea más fácil que te aprueben un crédito. ¿Cuáles son tus conclusiones?

```python
print("Crosstab Genero Vs Loan Status")
pd.crosstab(train_set.Gender,train_set.Loan_Status,margins=True)
#con los números sin %s es más dificil concluir, comentaremos resolviendo el BONUS
#ademas de que el número de hombres es mayor en la base
```

```python
print("BONUS : Crosstab Genero Vs Loan Status con % por filas")
pd.crosstab(train_set.Gender,train_set.Loan_Status).apply(lambda r: r/r.sum(), axis=1)
```

<span style="color:blue">No es una gran diferencia pero viendo las distribuciones por genero, los hombres tienen un % de créditos aprobados mayor que el de las mujeres; 70.5% Vs 62.7%</span>.

Haremos el mismo ejercicio ahora con la variable de historial de crédito.

```python
pd.crosstab(train_set.Credit_History,train_set.Loan_Status).apply(lambda r: r/r.sum(), axis=1)
```

<span style="color:blue">Definitivamente es más importante tener Historial de Crédito 90% de quienes no cuentan con el mismo son rechazados, vs un 79.2% de quienes si cuentan con uno. </span>.
